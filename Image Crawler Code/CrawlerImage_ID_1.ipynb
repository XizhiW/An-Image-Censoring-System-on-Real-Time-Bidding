```python
import os
import requests
total_num_of_images=665569
num_0=0
num_1=0
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/69.0.3497.100 Safari/537.36'}
root = r'E:\jupyter file\CrawlerImage\scrape_image_id_1\\'
for i in range(total_num_of_images):
    if status[i]==1:
        print('序号：')
        print(i)
        path=root+image_id[i]+'.jpg'
        try:
            if not os.path.exists(path):
                print(url[i])
                r = requests.get(url[i],headers=headers)
                print(r)
                with open(path, 'wb') as f:

                    f.write(r.content)
                    #f = r.replace(r, "1.jpg")
                    f.close()
                    print("文件保存成功")
                    num_0+=1
            else:
                print("文件已存在")
                print(i)
        except:
            print("爬取失败")
            num_1+=1
    
print("保存成功数:")
print(num_0)
print("保存失败数:")
print(num_1)
```
